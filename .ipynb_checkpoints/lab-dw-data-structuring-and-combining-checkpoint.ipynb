## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
df1
# df.info()
# df.dtypes
# df.nunique()
# df.describe()
# display(df.isna().sum())
# display((df.isna().sum()/df.shape[0])*100)
# df.head()
```

```python
# Step 2: Standardize column names
def clean_columns(df):
    df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]
    if 'st' in df.columns:
        df.rename(columns={'st': 'state'}, inplace=True)
    return df

df1 = clean_columns(df1)
df2 = clean_columns(df2)
df3 = clean_columns(df3)

# Step 3: Concatenate all dataframes
df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
df
```

```python
# Step 4: Clean 'gender' column (make consistent: 'M', 'F', None)
df['gender'] = df['gender'].astype(str).str.strip().str.upper().replace({'FEMALE': 'F', 'MALE': 'M', 'FEMAL': 'F', 'nan': None, 'NAN': None})
df.head(5)
```

```python
# Step 5: Clean 'customer_lifetime_value' column: remove '%' and convert to float
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(str).str.replace('%', '', regex=False)
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
df.head(5)
```

```python
# Step 6: Clean 'number_of_open_complaints' column: extract numeric value
df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(str).str.extract('(\d+)')
df['number_of_open_complaints'] = pd.to_numeric(df['number_of_open_complaints'], errors='coerce').fillna(0).astype(int)
df.head(5)
```

```python
display(df.isna().sum())
```

```python
# Step 7: Drop rows where all columns are NaN
df.dropna(inplace=True)

# Step 8: Reset index
df.reset_index(drop=True, inplace=True)

# Display cleaned dataframe
df
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
# Step 1: Load the marketing customer analysis data
marketing_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")

# marketing_df.info()
# print()
# marketing_df.dtypes
# print()
# marketing_df.nunique()
# print()
# marketing_df.describe()
# print()
# display(marketing_df.isna().sum())
# print()
# display((marketing_df.isna().sum()/marketing_df.shape[0])*100)
print()
# df.head()
display(marketing_df.columns)

# Step 2: Inspect and clean columns if needed
#marketing_df.columns = [c.strip().lower().replace(' ', '_') for c in marketing_df.columns]

# Step 2: Remove column 'unnamed:_0' that seems to be the same as the implecit index
marketing_df = marketing_df.drop(['unnamed:_0'], axis=1)
display(marketing_df)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Step 3: Total revenue by sales channel
revenue_by_channel = marketing_df.pivot_table(index='sales_channel', values='customer_lifetime_value', aggfunc='sum').round(2)

print(revenue_by_channel)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Step 4: Average customer lifetime value by gender and education
clv_by_gender_education = marketing_df.pivot_table(index='gender', columns='education', values='customer_lifetime_value', aggfunc='mean').round(2)

print(clv_by_gender_education)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here
# Step 5: Extract month from date
marketing_df['month'] = pd.to_datetime(marketing_df['effective_to_date']).dt.month

# Complaints by policy type and month, long format
complaints_summary = marketing_df.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum().reset_index()

print(complaints_summary)
```
